```python
import pandas as pd
import numpy as np
import spacy
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
df = pd.read_excel(r"C:\Users\sawanta\Downloads\News_Sentiment_Spacy\KeywordsandEntities_18112022.xlsx")
df.head()
```

```python
def spacy_entity(df):    
    df1 = nlp(df)
    df2 = [[w.text,w.label_] for w in df1.ents]
    return df2
df['NER'] = df['News'].apply(spacy_entity)
```

```python
def organisation(v):
    r = [(x,y) for x,y in v if ('ORG' == y)]
    return [(x) for x,y in r]

df['Entities'] = df['NER'].apply(lambda v: organisation(v))
df['Entities']= df['Entities'].apply(lambda x: list(set(x)) if x is not np.nan else np.nan)
```

```python
def person(v):
    r = [(x,y) for x,y in v if ('PERSON' == y)]
    return [(x) for x,y in r]

df['Person'] = df['NER'].apply(lambda v: person(v))
df['Person']= df['Person'].apply(lambda x: list(set(x)) if x is not np.nan else np.nan)
```

```python
df['News'] = df['News'].apply(lambda x: x.replace("._x000D_", " " ))
df['News'] = df['News'].apply(lambda x: x.replace("_x000D_", " " ))
df['News'] = df['News'].apply(lambda x: x.replace("&amp;quot;", " " ))
```
